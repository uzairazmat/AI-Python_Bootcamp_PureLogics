### a) Reshape into matrix form

```python
# TODO: Reshape and aggregate
temp_matrix = temp.reshape((10, 24))

wind_matrix = wind.reshape((10, 24))
# print(temp_matrix)
# Hint: temp_matrix = temp.reshape((10, 24))
# Write functions to find min, max, mean across rows

def find_min(matrix):
  return matrix.min(axis=1)

def find_max(matrix):
  return matrix.max(axis=1)

def find_mean(matrix):
  return matrix.mean(axis=1)

# Apply them to temp_matrix

min = find_min(temp_matrix)
min


```

### b) Normalize the temperature matrix
- Subtract the mean and divide by std deviation
- Do it manually using NumPy functions

```python
# TODO: Normalize temp_matrix
# Placeholder for function: def normalize(matrix):
def normalize(matrix):
  matrix_mean = matrix.mean(axis=0)  #why axis 0 b/c it calculate row wise same hour for all days
  matrix_std = matrix.std(axis=0)
  return (matrix - matrix_mean) / matrix_std

#
# return ...

# Apply it to temp_matrix

normalized_temp = normalize(temp_matrix)
normalized_temp
```

## Final Challenge: Write Your Own Function
Write a function `daily_summary(matrix)` that takes a NumPy matrix of shape (10, 24) and returns a summary dictionary for each day.

```python
# TODO: Write and test your func
def daily_summary(matrix):
  if matrix.shape != (10, 24): # handling error excetptions due to given condition
    raise ValueError("Input matrix should have shape (10, 24)")
  matrix_mean = find_mean(matrix) # function call
  matrix_max = find_max(matrix) # function call
  matrix_min = find_min(matrix) # function call
  summary_dict = {}
  for i in range(10):
    summary_dict[f"Day {i}"] = {  #key that represent every day with numbers
      "mean": matrix_mean[i],
      "max": matrix_max[i],
      "min": matrix_min[i]
    }
  return summary_dict

# Example usage:
summaries = daily_summary(temp_matrix)
print(summaries)
```
